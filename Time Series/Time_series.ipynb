## overview
A time series is a sequence of data points collected or recorded at successive points in time, typically at uniform intervals.

Examples : daily stock prices, monthly sales figures, and hourly temperature readings.
Time series analysis use in:
* Forecasting
* Anomaly detection
* Understanding patterns
* Improving efficiency
* Risk management
 ## Components of time series
Time series data can be broken down into several key components, each of which helps in understanding the underlying patterns and structures. The primary components of a time series are:
* Trend
* Seasonality
* Cyclical
* Irregularity

### **Trend:**
The tendency of a time series data to increase or decrease during a long period of time.
### **Seasonality:**
the variation in the data that have a regular and periodic manner over a span of less than a year.
### **Cyclical:**
the oscillatory movements in the time series data that happen in more than a year.
### **Irregularity:**
Irregular or unpredictable fluctuations that can obscure the underlying patterns.



## Additive vs Multiplicative
Additive and multiplicative decomposition are two common approaches used to separate a time series into its constituent components: trend, seasonality, cyclical and irregularity. The choice between these two methods often depends on the nature of the data.
### Additive decomposition
implies that time-series data is a function of the sum of its components. This can be represented with the following equation:

\begin{align}
\text{Y(t) = T(t) + S(t) + C(t) + I(t)}
\end{align}

y = output

T = Trend

S = Seasonality

C = Cyclical

I = Irregularity

### Multiplicative decomposition
define temporal data as a function of the product of its components. In the form of an equation:

\begin{align}
\text{Y(t) = T(t)} × \text{S(t)} × \text{C(t)} × \text{I(t)}
\end{align}


## Stationary vs non-stationary
A time series has stationarity when the observations are not dependent on the time. Statistical properties of these time series will not change with time thus they will have constant mean, variance, and covariance.

Stationarity is an important property because many time series models, such as ARIMA, assume that the series being modeled is stationary. the Augmented Dickey-Fuller (ADF) test or the KPSS test can be used to formally test for stationarity.

If a time series is not stationary, it can often be transformed into a stationary series by differencing, detrending , or taking logarithms.


### Augmented Dickey-Fuller (ADF) test
Augmented Dickey-Fuller (ADF) Test or Unit Root Test
The ADF test is the most popular statistical test. It is done with the following assumptions:

**Null Hypothesis (HO):** Series is non-stationary

**Alternate Hypothesis (HA):** Series is stationary

* P-value > 0.05 fail to reject (H0)

* P-value <= 0.05 accept (H1)

### Autocorrelation Function (ACF)
The ACF measures the correlation between a time series and a lagged version of itself.
It helps us understand how current values of the series are related to its past values.

This is useful for identifying patterns, trends, or repeated cycles in the data.


### Partial Autocorrelation Function (PACF)
The Partial Autocorrelation Function (PACF) is a key concept in time series analysis, particularly in the context of identifying the appropriate parameters for autoregressive models like ARIMA. Here’s a breakdown of what PACF is and how it is used:

The PACF measures the correlation between a time series and its lagged values, while controlling for the values of the series at shorter lags. In simpler terms, it tells you how much a variable is related to its past values after removing the influence of the intervening lags.

This is useful for identifying patterns, trends, or repeated cycles in the data.


## Common Types of Time Series Forecasting Models
### **ARIMA (AutoRegressive Integrated Moving Average):**
Combines autoregression, differencing (to make the series stationary), and moving averages.

Suitable for univariate time series without seasonal patterns.
### **SARIMA (Seasonal ARIMA):**
An extension of ARIMA that accounts for seasonality.
It adds seasonal components to the model.
### **Exponential Smoothing:**
A family of forecasting methods that weigh past observations with exponentially decreasing weights.

Simple Exponential Smoothing, Holt’s Linear Trend Model, and Holt-Winters Seasonal Model are common variations.
### **Prophet:**
Developed by Facebook, this model is designed to handle seasonality and holidays effectively.

It’s user-friendly and works well with missing data.
### **State Space Models:**
These models describe the observed data as a combination of unobserved states.
They can handle irregularities and are often used in conjunction with the Kalman filter.
### **Machine Learning Approaches:**
Techniques like Random Forests, XGBoost, or Neural Networks (e.g., LSTM) can be applied to time series data.

These methods can capture complex nonlinear relationships in data.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.stattools import adfuller

# Generate synthetic time series data
np.random.seed(0)
time = np.arange(100)
data = 10 + 0.5 * time + np.sin(time / 3) + np.random.normal(size=time.size)

# Create a DataFrame
df = pd.DataFrame(data, columns=['Value'])
df['Time'] = pd.date_range(start='2020-01-01', periods=len(df), freq='D')
df.set_index('Time', inplace=True)

# Plot the data
plt.figure(figsize=(12, 6))
plt.plot(df, label='Original Data')
plt.title('Synthetic Time Series Data')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()
plt.show()

# Check for stationarity
result = adfuller(df['Value'])
print('ADF Statistic:', result[0])
print('p-value:', result[1])

# Fit ARIMA model
model = ARIMA(df['Value'], order=(1, 1, 1))  # Adjust the order as needed
model_fit = model.fit()

# Summary of the model
print(model_fit.summary())

# Forecasting the next 10 periods
forecast = model_fit.forecast(steps=10)
forecast_index = pd.date_range(start=df.index[-1] + pd.Timedelta(days=1), periods=10)
forecast_series = pd.Series(forecast, index=forecast_index)

# Plot the forecast
plt.figure(figsize=(12, 6))
plt.plot(df, label='Original Data')
plt.plot(forecast_series, label='Forecast', color='red')
plt.title('Time Series Forecasting with ARIMA')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.stattools import adfuller

# Generate synthetic time series data
np.random.seed(0)
time = np.arange(100)
data = 10 + 0.5 * time + np.sin(time / 3) + np.random.normal(size=time.size)

# Create a DataFrame
df = pd.DataFrame(data, columns=['Value'])
df['Time'] = pd.date_range(start='2020-01-01', periods=len(df), freq='D')
df.set_index('Time', inplace=True)

# Plot the original data
plt.figure(figsize=(12, 6))
plt.plot(df, label='Original Data')
plt.title('Synthetic Time Series Data')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()
plt.show()

# Check for stationarity
result = adfuller(df['Value'])
print('ADF Statistic:', result[0])
print('p-value:', result[1])

# Make the series stationary by differencing
df['Differenced'] = df['Value'].diff().dropna()

# Drop the NaN values created by differencing
df_diff = df['Differenced'].dropna()

# Check for stationarity on the differenced series
result_diff = adfuller(df_diff)
print('ADF Statistic (Differenced):', result_diff[0])
print('p-value (Differenced):', result_diff[1])

# Fit ARIMA model on the differenced series
model = ARIMA(df['Value'], order=(1, 1, 1))  # Keep the differencing (d=1)
model_fit = model.fit()

# Summary of the model
print(model_fit.summary())

# Forecasting the next 10 periods
forecast = model_fit.forecast(steps=10)
forecast_index = pd.date_range(start=df.index[-1] + pd.Timedelta(days=1), periods=10)
forecast_series = pd.Series(forecast, index=forecast_index)

# Plot the original data and the forecast
plt.figure(figsize=(12, 6))
plt.plot(df['Value'], label='Original Data')
plt.plot(forecast_series, label='Forecast', color='red')
plt.title('Time Series Forecasting with ARIMA')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.stattools import adfuller

# Step 1: Fetch Bitcoin price data
btc_data = yf.download('BTC-USD', start='2022-01-01', end='2024-01-01', interval='1d')
btc_data = btc_data['Close']  # Use the closing prices

# Step 2: Visualize the data
plt.figure(figsize=(12, 6))
plt.plot(btc_data, label='BTC-USD Closing Price')
plt.title('Bitcoin Price History')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.show()

# Step 3: Check for stationarity
result = adfuller(btc_data.dropna())
print('ADF Statistic:', result[0])
print('p-value:', result[1])

# Step 4: Fit ARIMA model
model = ARIMA(btc_data, order=(5, 1, 0))  # You can tune the order based on ACF and PACF plots
model_fit = model.fit()

# Step 5: Summary of the model
print(model_fit.summary())

# Step 6: Forecasting the next 30 days
forecast = model_fit.forecast(steps=30)
forecast_index = pd.date_range(start=btc_data.index[-1] + pd.Timedelta(days=1), periods=30)
forecast_series = pd.Series(forecast, index=forecast_index)

# Step 7: Plot the forecast
plt.figure(figsize=(12, 6))
plt.plot(btc_data, label='BTC-USD Closing Price')
plt.plot(forecast_series, label='Forecast', color='red')
plt.title('Bitcoin Price Forecast')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.show()
```

### predict bitcoin price with Exponential Smoothing model(in 30 days later):

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
from statsmodels.tsa.holtwinters import ExponentialSmoothing

# Step 1: Fetch Bitcoin price data
btc_data = yf.download('BTC-USD', start='2022-01-01', end='2024-01-01', interval='1d')
btc_data = btc_data['Close']  # Use the closing prices

# Step 2: Visualize the data
plt.figure(figsize=(12, 6))
plt.plot(btc_data, label='BTC-USD Closing Price')
plt.title('Bitcoin Price History')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.show()

# Step 3: Fit Exponential Smoothing model
model = ExponentialSmoothing(btc_data, trend='add', seasonal='add', seasonal_periods=30)
model_fit = model.fit()

# Step 4: Forecasting the next 30 days
forecast = model_fit.forecast(steps=30)
forecast_index = pd.date_range(start=btc_data.index[-1] + pd.Timedelta(days=1), periods=30)
forecast_series = pd.Series(forecast, index=forecast_index)

# Step 5: Plot the forecast
plt.figure(figsize=(12, 6))
plt.plot(btc_data, label='BTC-USD Closing Price')
plt.plot(forecast_series, label='Forecast', color='red')
plt.title('Bitcoin Price Forecast with Exponential Smoothing')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.show()
```



### predict bitcoin price with Prophet model (in 30 days later):

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
from prophet import Prophet

# Step 1: Fetch Bitcoin price data
btc_data = yf.download('BTC-USD', start='2022-01-01', end='2024-01-01', interval='1d')
btc_data = btc_data[['Close']].reset_index()
btc_data.columns = ['ds', 'y']  # Rename columns for Prophet

# Step 2: Visualize the data
plt.figure(figsize=(12, 6))
plt.plot(btc_data['ds'], btc_data['y'], label='BTC-USD Closing Price')
plt.title('Bitcoin Price History')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.show()

# Step 3: Fit Prophet model
model = Prophet()
model.fit(btc_data)

# Step 4: Make future dataframe for the next 30 days
future = model.make_future_dataframe(periods=30)
forecast = model.predict(future)

# Step 5: Plot the forecast
fig = model.plot(forecast)
plt.title('Bitcoin Price Forecast with Prophet')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.show()
```
